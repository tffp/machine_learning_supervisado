- Sacar la media de cada continente,listar todos los continentes por su media en los datos "Poblacion_Paises.txt"

```python
df = pd.read_csv("Poblacion_Paises.txt", encoding = "utf-8") 
df.head(10)
```

```python
media_asia  = df[df['Continente']=='Asia']['Poblaci√≥n'].mean()
media_Africa =  df[df['Continente']=='Africa']['Poblaci√≥n'].mean()
media_Americas = df[df['Continente']=='Americas']['Poblaci√≥n'].mean()
media_Europe = df[df['Continente']=='Europe']['Poblaci√≥n'].mean()
media_Oceania = df[df['Continente']=='Oceania']['Poblaci√≥n'].mean()
m = [media_asia,media_Africa,media_Americas,media_Europe,media_Oceania]
```

```python
medias = pd.DataFrame(m,['Asia','Africa','Americas','Europe','Oceania'], columns = ['Medias'])
medias
```

- Listar todos los pa√≠ses en orden de crecimiento (en porcentaje) en el intervalo de 2002 a 2007

```python
df_2002 = df[df['A√±o']==2002]
df_2007 = df[df['A√±o']==2007]
# con el drop= True para eliminar el index  previo, al concatenar las tablas
df_ext = pd.concat([df_2002.reset_index(drop=True), df_2007.reset_index(drop=True)], axis=1, ignore_index=True)
df_ext.drop(columns=[4,5], inplace=True)
df_ext.columns = ['Pa√≠s','Continente','A√±o 2002','Poblaci√≥n 2002','A√±o 2007','Poblaci√≥n 2007']
df_ext
```

```python
df_ext['Crecimiento'] = df_ext['Poblaci√≥n 2007']-df_ext['Poblaci√≥n 2002']
df_ext.sort_values(by=['Crecimiento'], ascending=False)
```

```python
df_ext['crecimiento_respecto_07'] = df_ext['Crecimiento'] / df_ext['Poblaci√≥n 2007']
df_ext['%_07/02'] = df_ext['Poblaci√≥n 2007'] / df_ext['Poblaci√≥n 2002']
df_ext
```
